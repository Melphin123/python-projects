```python
def choice():
    choice = input("Enter the choice odd/even: ")
    global odd
    global even
    if choice == 'even':
        even = 'computer' 
        odd ='player'
        print("computer will be even and player will be odd")
    elif choice == 'odd':
        odd = 'computer' 
        even = 'player'
        print("computer will be odd and player will be even")
    return odd,even
    
    
```
